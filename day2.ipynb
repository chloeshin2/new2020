```

W = tf.Variable(3.5)  #W에 임의의 값을 지정
b = tf.Variable(2.5)  #b에 임의의 값을 지정
```

```
#XOR 
x = np.array([[1,1],[1,0],[0,1],[0,0]])
y = np.array([[0],[1],[1],[0]])
#앤드 연산을 통해서, 

#y = ax +b 
#퍼셉트론을 중간에 넣을것인데...
# weight, bias ? as intercept

#y = W*x + b  (공식화한 함수모양)
#행렬식이라서 대문자로 한다.. 얘를 처리할 행렬도 같은 크기여야해서 대문자 처리로 행렬임을 보여준다

#선형회귀방식과 유사, 기울기와 절편을 구하는것. 

W = tf.Variable(3.5)  #W에 임의의 값을 지정
b = tf.Variable(2.5)  #b에 임의의 값을 지정

# 0 = 3.5 * 1 + 2.5
# 0 = 6 실제 값고 예측값이 안맞는것은 오차고 오차를 그래프에서 가장낮은 접점을 찾는게 머신러닝

#안맞는게 맞아요ㅋㅋ

#오차를 최소화하는게 목적 !!(***)

learning_rate = 0.01  #머신러닝 학습율

for i in range(2001):
 with tf.GradientTape() as tape:
    model = W*x +b
    cost = tf.reduce_mean(tf.square(model - y))     #줄여가는 함수 

 w1, b1 = tape.gradient(cost, [W, b])
 W.assign_sub(learning_rate*w1)
 b.assign_sub(learning_rate*b1)

 predicted = tf.cast(model>0.5, dtype = tf.float32)
 accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, y), dtype = tf.float32))

 if i%10 == 0:          #i는 횟수고 10번까지만, i값을 10으로 나눠서, 0이될때까지? 
  print( i, W.numpy(), b.numpy(), cost)

print("모델: ", model)
print("예측값:" , predicted)
print("정확도:", accuracy)
#값이 작아지면 

```

```
model = tf.keras.Sequential([
        tf.keras.layers.Dense(units=8, activation ='sigmoid', input_shape=(2,)),
        tf.keras.layers.Dense(units=1, activation = 'sigmoid')
])

```

```
model.compile(optimizer= tf.keras.optimizers.SGD(learning_rate=0.1), loss ="binary_crossentropy",metrics =['accuracy'])
#손실 함수의 종류도 추후에.. 정확도
```

```
model.fit(x,y, epochs=3000, batch_size=1)
model.summary()
model.predict(x)

```
